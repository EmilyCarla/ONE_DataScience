```python
#importando todas as bibliotecas necessarias
import pandas as pd
import requests
import json
```

```python
# carregando os dados da API
dados_usuarios = requests.get('https://jsonplaceholder.typicode.com/users')
resultado = json.loads(dados_usuarios.text)
pd.DataFrame(resultado)
```

```python
# normalizando os dados
dados_usuarios_normalizado = pd.json_normalize(resultado, sep='_')
dados_usuarios_normalizado
```

BANCO DE DADOS

```python
import sqlalchemy
```

```python
from sqlalchemy import create_engine, MetaData, Table, inspect
```

```python
engine = create_engine('sqlite:///:memory:')
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/refs/heads/main/clientes_banco.csv'
```

```python
import pandas as pd
```

```python
dados = pd.read_csv(url)

dados.head()
```

```python
dados.to_sql('clientes', engine, index=False)
```

```python
inspector = inspect(engine)
```

```python
print(inspector.get_table_names()) #devolve o nome das tabelas contidas no banco
```

```python
query = 'SELECT * FROM clientes WHERE Categoria_de_renda = "Empregado"'
```

```python
empregados = pd.read_sql(query, engine)
empregados
```

```python
empregados.to_sql('empregados', con=engine, index=False) # con = especifica o banco de dados
```

```python
pd.read_sql_table('empregados', engine)
```

read_sql() => Lê uma consulta
read_sql_table() -> Lê uma tabela inteira

```python
pd.read_sql_table('empregados', engine, columns=['ID_Cliente', 'Grau_escolaridade', 'Rendimento_anual'])
```

```python
from sqlalchemy import text
```

```python
query = 'DELETE FROM clientes WHERE ID_Cliente=5008804'
with engine.connect() as conn: #with garante que a conexão com o BD seja fechada após a consulta
    result = conn.execute(text(query))
    conn.commit()
```

```python
pd.read_sql_table('clientes', engine)
```

```python
query = 'UPDATE clientes SET Grau_escolaridade="Ensino superior" WHERE ID_Cliente=5008808'
with engine.connect() as conn:
    result = conn.execute(text(query))
    conn.commit()
```

```python
pd.read_sql_table('clientes', engine)
```
