Missão 1: Time de vendas

Construir uma visualização para indicar os 10 maiores clientes por total de venda.

```python
df_cliente_vendas = df.groupby(['nome_cliente'])['vendas'].sum().nlargest(10) #nlargest() -> ordena os valores e receb o parametro de qnts valores são
df_cliente_vendas = df_cliente_vendas.reset_index()
df_cliente_vendas
```

```python
#ajustando para o index começar em 1
df_cliente_vendas.columns = ['Clientes', 'Vendas']
df_cliente_vendas['Ranque'] = df_cliente_vendas.index+1 #cria a coluna "ranque" com o valor dos indices
df_cliente_vendas.set_index('Ranque', inplace=True) #
df_cliente_vendas
```

```python
s = df_cliente_vendas.style
s
```

```python
s.format({'Vendas': 'R$ {:,.2f}'}) # Formatação para deixar em formato de reais
s
```

# Missão 2: Time comercial
Constuir uma visualização que permita associar o total ganho em vendas e o total lucrado por tipo de produto

```python
# organizar os dados
df_venda_lucro = df.groupby(['tipo_produto'])[['vendas', 'lucro']].sum()
df_venda_lucro.index.name = 'tipo produto'
df_venda_lucro
```

```python
estilo_produto = df_venda_lucro.style
estilo_produto
```

```python
#formatar os valores float
estilo_produto.format('R$ {:,.2f}')\
              .highlight_max(color='lightgreen')\
              .highlight_min(color='#F16165')
estilo_produto
```

```python
# usando gradiente
estilo_produto.format('R$ {:,.2f}').background_gradient(cmap='Greens')
estilo_produto
```

```python
#padronizando a tabela
cabecalho = {
    'selector': 'th', #pega o cabeçalho da tabela
    'props': 'font_weight: bold; font-family: Arial; text-align: center; text-transform: captalize;'
}
estilo_produto.set_table_styles([cabecalho], overwrite=False)
```

# Missão 3: Time de logística
Construir visualização que permita entender a distribuição de pedidos por região do Brasil

```python
df_regiao = pd.DataFrame(df['regiao'].value_counts())
df_regiao.columns = ['N° pedidos']
df_regiao
```

```python
df_regiao.index.name = 'Regiao'
```

```python
import numpy as pn
```

```python
porcentagem = df_regiao['N° pedidos'].to_numpy()
porcentagem = 100 * porcentagem / porcentagem.sum()
porcentagem
```

```python
df_regiao['Porcentagem regiao'] = porcentagem
df_regiao
```

```python
estilo_regiao = df_regiao.style
estilo_regiao
```

```python
cabecalho = {
    'selector': 'th',
    'props' : 'font-weight: bold; font-family: Arial; text-align: right; background-color: black'
}
celulas = {
    'selector' : 'td', #se refer as celulas
    'props' : 'background-color: black;'

}

estilo_regiao.set_table_styles([cabecalho, celulas])
```

```python
estilo_regiao.format({'Porcentagem regiao' : '{:.2f} %'})\
              .bar(subset='Porcentagem regiao', vmin=0, vmax=100.0, color = '#9CD33B')
```
