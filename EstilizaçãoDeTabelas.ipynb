Missão 1: Time de vendas

Construir uma visualização para indicar os 10 maiores clientes por total de venda.

```python
df_cliente_vendas = df.groupby(['nome_cliente'])['vendas'].sum().nlargest(10) #nlargest() -> ordena os valores e receb o parametro de qnts valores são
df_cliente_vendas = df_cliente_vendas.reset_index()
df_cliente_vendas
```

```python
#ajustando para o index começar em 1
df_cliente_vendas.columns = ['Clientes', 'Vendas']
df_cliente_vendas['Ranque'] = df_cliente_vendas.index+1 #cria a coluna "ranque" com o valor dos indices
df_cliente_vendas.set_index('Ranque', inplace=True) #
df_cliente_vendas
```
